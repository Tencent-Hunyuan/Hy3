# Example 5: Reasoning Mode —— 推理模式对比

对比 Hy3 的三种 reasoning mode：
- `no_think`：直接回答（快思考）
- `low`：轻量推理
- `high`：深度推理（慢思考）

```python
import time
from openai import OpenAI

API_KEY = "sk-你的APIKey"  # TODO: 替换

client = OpenAI(
    base_url="https://tokenhub.tencentmaas.com/v1",
    api_key=API_KEY,
)

prompt = """有三个盒子：一个只装苹果，一个只装橘子，一个既装苹果又装橘子。
所有标签都贴错了。从一个盒子里拿出一个水果看，最少需要拿几次？为什么？"""
```

## 1. no_think（快思考）

```python
print("Reasoning Mode: no_think（快思考）\n")

start = time.time()
response = client.chat.completions.create(
    model="hy3", messages=[{"role": "user", "content": prompt}],
    temperature=0.7, top_p=1.0,
    extra_body={"reasoning_effort": "no_think"},
)
elapsed = time.time() - start

msg = response.choices[0].message
print(f"耗时: {elapsed:.2f}s")
print(f"reasoning_content: {'无' if not getattr(msg, 'reasoning_content', None) else '有'}")
print(f"\n回答:\n{msg.content[:300]}")
```

## 2. low（轻量推理）

```python
print("Reasoning Mode: low（轻量推理）\n")

start = time.time()
response = client.chat.completions.create(
    model="hy3", messages=[{"role": "user", "content": prompt}],
    temperature=0.7, top_p=1.0,
    extra_body={"reasoning_effort": "low"},
)
elapsed = time.time() - start

msg = response.choices[0].message
print(f"耗时: {elapsed:.2f}s")
print(f"reasoning_content: {'有' if getattr(msg, 'reasoning_content', None) else '无'}")
if getattr(msg, 'reasoning_content', None):
    print(f"\n思考过程片段:\n{msg.reasoning_content[:200]}...")
print(f"\n最终回答:\n{msg.content[:300]}")
```

## 3. high（慢思考）

```python
print("Reasoning Mode: high（慢思考）\n")

start = time.time()
response = client.chat.completions.create(
    model="hy3", messages=[{"role": "user", "content": prompt}],
    temperature=0.7, top_p=1.0,
    extra_body={"reasoning_effort": "high"},
)
elapsed = time.time() - start

msg = response.choices[0].message
print(f"耗时: {elapsed:.2f}s")
print(f"reasoning_content: {'有' if getattr(msg, 'reasoning_content', None) else '无'}")
if getattr(msg, 'reasoning_content', None):
    print(f"\n思考过程片段:\n{msg.reasoning_content[:300]}...")
print(f"\n最终回答:\n{msg.content[:300]}")
```

## 对比总结

| 模式 | reasoning_content | 速度 | 适用场景 |
|------|:-:|:----:|---------|
| no_think | ❌ 无 | 最快 | 日常对话、简单问答 |
| low | ✅ 简短 | 较快 | 中等复杂度任务 |
| high | ✅ 详细 | 较慢但准确 | 数学/编码/逻辑推理 |
