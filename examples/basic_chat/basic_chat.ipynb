# Hy3 API - 基础聊天示例

本示例演示如何使用 Hy3 API 进行单轮和多轮对话。

```python
from openai import OpenAI
import os
from dotenv import load_dotenv
load_dotenv()
client = OpenAI(
    api_key=os.getenv("API_KEY", "EMPTY"),
    base_url=os.getenv("BASE_URL", "http://127.0.0.1:8000/v1")
)
```

## 单轮对话

```python
response = client.chat.completions.create(
    model="hy3",
    messages=[
        {"role": "user", "content": "你好！请简单介绍一下你自己。"},
    ],
    temperature=0.9,
    top_p=1.0,
    extra_body={"reasoning_effort": "low"},
)

print(f"响应 ID: {response.id}")
print(f"模型: {response.model}")
print(f"生成时间: {response.created}")
print(f"\n回复内容:")
print(response.choices[0].message.content)
```

## Response 解析

```python
print("=== Response 结构 ===")
print(f"id: {response.id}")
print(f"object: {response.object}")
print(f"created: {response.created}")
print(f"model: {response.model}")

print("\n=== Choices ===")
for i, choice in enumerate(response.choices):
    print(f"  [{i}] finish_reason: {choice.finish_reason}")
    print(f"      message.role: {choice.message.role}")
    print(f"      message.content: {choice.message.content[:50]}...")

print("\n=== Usage ===")
print(f"  prompt_tokens: {response.usage.prompt_tokens}")
print(f"  completion_tokens: {response.usage.completion_tokens}")
print(f"  total_tokens: {response.usage.total_tokens}")
```

## 多轮对话

```python
messages = []

# 第一轮
user_msg_1 = "什么是 MoE 模型？"
messages.append({"role": "user", "content": user_msg_1})

response_1 = client.chat.completions.create(
    model="hy3",
    messages=messages,
    temperature=0.9,
    top_p=1.0,
    extra_body={"reasoning_effort": "low"},
)
assistant_msg_1 = response_1.choices[0].message.content
messages.append({"role": "assistant", "content": assistant_msg_1})

print(f"User: {user_msg_1}")
print(f"Assistant: {assistant_msg_1}")
print("-" * 50)
```

```python
# 第二轮
user_msg_2 = "Hy3 在 MoE 架构上有什么创新？"
messages.append({"role": "user", "content": user_msg_2})

response_2 = client.chat.completions.create(
    model="hy3",
    messages=messages,
    temperature=0.9,
    top_p=1.0,
    extra_body={"reasoning_effort": "low"},
)
assistant_msg_2 = response_2.choices[0].message.content

print(f"User: {user_msg_2}")
print(f"Assistant: {assistant_msg_2}")
```

## 多轮对话关键要点

1. **维护完整历史**：每次请求必须包含所有历史消息
2. **角色顺序**：`user` 和 `assistant` 交替出现
3. **上下文限制**：总 token 数不能超过模型限制（256k）
