# 05 Reasoning Mode：思考过程开/关对比

演示如何开启/关闭 Hy3 的思考模式，并读取 `reasoning_content`。

```python
import os
from openai import OpenAI

client = OpenAI(
    api_key=os.environ.get("HY3_API_KEY"),
    base_url="https://tokenhub.tencentmaas.com/v1",
)

PROMPT = "9.11 和 9.8 哪个更大？请说明原因。"
```

```python
def ask_with_thinking(enabled: bool):
    extra_body = {"thinking": {"type": "enabled" if enabled else "disabled"}}
    response = client.chat.completions.create(
        model="hy3",
        messages=[{"role": "user", "content": PROMPT}],
        temperature=0.3,
        max_tokens=512,
        extra_body=extra_body,
    )

    msg = response.choices[0].message
    print(f"=== {'开启思考' if enabled else '关闭思考'} ===")
    reasoning = getattr(msg, "reasoning_content", None)
    if reasoning:
        print("【思考过程】")
        print(reasoning)
        print("\n【最终回答】")
    print(msg.content)
    print()

ask_with_thinking(False)
ask_with_thinking(True)
```
