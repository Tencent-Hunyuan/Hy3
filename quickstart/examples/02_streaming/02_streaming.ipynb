# Example 2: Streaming（流式请求）

展示流式响应的处理方式。

！注意配置好.env

1.引入工具

```python
from openai import OpenAI
from dotenv import load_dotenv
import os
```

2.导入配置

```python
load_dotenv()

client = OpenAI(
    api_key=os.getenv("HY3_API_KEY"),
    base_url="https://tokenhub.tencentmaas.com/v1",
)
```

3.流式请求

流式请求返回的是多个chunk块，通过这种将一段长话分短块返回的方式缩短用户等待时间

这里的配置主要是stream=True

include_usage的作用是对用量进行统计，因为使用流式输出最后一般不会告诉使用了多少token，因为多个chunk里没有汇总信息，所以需要使用这个在最后一个chunk强制带上总使用量

```python
# 流式请求
print("=== 流式请求示例 ===")
stream = client.chat.completions.create(
    model="hy3",
    messages=[
        {"role": "user", "content": "请写一段关于人工智能发展的简短介绍。"},
    ],
    stream=True,
    stream_options={"include_usage": True},
)

print("\n=== 流式输出 ===")
full_content = ""
for chunk in stream:
    if chunk.choices and chunk.choices[0].delta.content:
        content = chunk.choices[0].delta.content
        full_content += content
        print(content, end="", flush=True)
    if chunk.usage:
        print("\n\n=== Token 使用 ===")
        print(f"输入: {chunk.usage.prompt_tokens} tokens")
        print(f"输出: {chunk.usage.completion_tokens} tokens")
        print(f"总计: {chunk.usage.total_tokens} tokens")

print("\n\n=== 完整响应 ===")
print(full_content)
```
